```python
df_1 = pd.read_csv('index_finger_data.csv')
df_1
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df_1['timestamp_ms'], df_1['ch1'], label='Channel 1')
plt.plot(df_1['timestamp_ms'], df_1['ch2'], label='Channel 2')
plt.xlabel('time(in ms)')
plt.ylabel('EMG Signal')
plt.title('(Channel 1 and Channel 2) v/s (Time)')
plt.legend()
plt.show()


```
